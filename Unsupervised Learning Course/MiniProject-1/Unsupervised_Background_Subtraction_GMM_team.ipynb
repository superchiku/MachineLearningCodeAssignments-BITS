```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image = plt.imread('deer.jpg')/255

plt.figure(figsize = (15,8))
plt.imshow(image)
```

```python
from sklearn.mixture import GaussianMixture as GMM
```

```python
#3 dimension to 2 dimension
img2 = img.reshape((-1,3))
print(img2)
```

```python
print(img2.shape)
```

```python
gmm_model = GMM(n_components = 3, covariance_type = 'tied')
gmm_model.fit(img2)
gmm_labels = gmm_model.predict(img2)
```

```python
print(gmm_model)
```

```python
print(gmm_labels)
print(gmm_labels.shape)
```
