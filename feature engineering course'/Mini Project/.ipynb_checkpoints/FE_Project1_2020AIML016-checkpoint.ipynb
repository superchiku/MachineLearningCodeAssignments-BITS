```python
####### Approach Explanation #######
'''
Answering each question with the code implmentation in each cell 

'''


########Load Libraries#######

import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.preprocessing import StandardScaler # using standard scaler as data for most features is normally distributed
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.manifold import TSNE
from sklearn.model_selection import StratifiedKFold
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.feature_selection import RFECV
from tqdm import tqdm 
from math import sqrt 
from time import gmtime, strftime

%matplotlib inline

######## Capture the start time to check the runt ime of the whole notebook #########
startTime = strftime("%a, %d %b %Y %H:%M:%S +0000", gmtime())
```

```python

####################################
#          Pre-Process Data        #
####################################

######## Load Data ########

BASE_PATH = './'
os.listdir(BASE_PATH)

#df_load_features = pd.read_csv(os.path.join(BASE_PATH, 'adult.data'),skipinitialspace=True)
df_whole_XLS = pd.ExcelFile(os.path.join(BASE_PATH, 'HR-Records_Final.xlsx'))
df_appdata = pd.read_excel(df_whole_XLS, 'Application data')
df_canddata = pd.read_excel(df_whole_XLS, 'Candidate Data')
df_interdata = pd.read_excel(df_whole_XLS, 'Interview Data ')
df_load_features = pd.merge(df_appdata, df_canddata, on="Candidate ID")
df_load_features = pd.merge(df_load_features, df_interdata, on="Candidate ID")
#df_load_features.to_csv("MergedHRFinalData.csv")

print("\ndata types of dataframe\n",df_load_features.dtypes)



######## Replace invalid character ? ########

df_load_features.replace("?", np.nan, inplace = True)

######## print quality of data ########

print("\n\nHead data \n\n",df_load_features.head(20))
print("\n\nTail data \n\n",df_load_features.tail(20))
print("\n\nDescribe data \n\n",df_load_features.describe())
print("\n\nNull values present:  ",df_load_features.isnull().values.any())
print("\n\nLength of dataframe:",len(df_load_features))

######## Drop duplicate rows whereever found  ########

duplicate_df = df_load_features[df_load_features.duplicated()] 
  
print("\n\nDuplicate Rows :\n\n" , duplicate_df) 
print("\n\ncount of duplicate rows :" , len(duplicate_df))

df_load_features.drop_duplicates(keep=False,inplace=True) 
print("\n\nlength of dataframe after dropping duplicates:",len(df_load_features))

duplicate_df = df_load_features[df_load_features["Candidate ID"].duplicated()] #Check whether cadidate id is duplicated
print("\n\ncount of duplicate Candidate ID rows :" , len(duplicate_df))

######## Manage missing values and replace by mode to keep the data consistent ########

for column in df_load_features:
    print("\n\nmode of column: {} is {}".format(column,df_load_features[column].value_counts().idxmax()))
    df_load_features[column].replace(np.nan, df_load_features[column].value_counts().idxmax(), inplace = True)
print("\n\nAll misisng values if any replaced by mode")
print("\n\nIs any null value present after replacing the missing values:  ",df_load_features.isnull().values.any())


######## Drop na columns if still present #########

df_load_features = df_load_features.dropna()

######## Data transformation to set the dataframe in the right format #########

df_load_features_raw = df_load_features.copy() # Store non transformed data in another dataframe for future use


df_load_features['ElapsedDays_Application_Joining'] = df_load_features['Date of Joining'].sub(df_load_features[
    'Date of Application'], axis=0) #Use elapsed days between date of joining and date of application

df_load_features['ElapsedDays_Application_Joining'] = df_load_features['ElapsedDays_Application_Joining'] / np.timedelta64(1, 'D')
df_load_features['Hike Difference'] = df_load_features['Percentage hike offered (CTC)'].sub(
    df_load_features['Percentage hike (CTC) expected '],axis=0)*100


df_load_features['ElapsedDays_InterviewDates'] = df_load_features['Date of last Interview'].sub(df_load_features[
    'Written Test Date'], axis=0) #Use elapsed days between date of both interviews

df_load_features['ElapsedDays_InterviewDates'] = df_load_features['ElapsedDays_InterviewDates'] / np.timedelta64(1, 'D')

df_load_features.drop("Candidate ID", axis = 1, inplace=True) #Candidate ID is not required
df_load_features.drop("Current organization", axis = 1, inplace=True) #Current Organization is not required
df_load_features.drop("Date of Application", axis = 1, inplace=True) #Drop date of application as we dont need it
df_load_features.drop("Date of Interview", axis = 1, inplace=True) #Drop Date of Interview as we dont need it
df_load_features.drop("Date of Job Offer", axis = 1, inplace=True) #Drop Date of Job Offer as we dont need it
df_load_features.drop("Date of Offer extension", axis = 1, inplace=True) #Drop Date of Offer extension as we dont need it
df_load_features.drop("Date of Joining", axis = 1, inplace=True) #Drop Date of Joining as we dont need it
df_load_features.drop("Offered Salary", axis = 1, inplace=True) #Drop offered salary as we have offered band
df_load_features.drop("Current CTC", axis = 1, inplace=True) #Drop current CTC as we have Hike Difference
df_load_features.drop("Current gross salary", axis = 1, inplace=True) #Drop Current gross salary as we have Hike Difference
df_load_features.drop("Expected CTC", axis = 1, inplace=True) #Drop Expected CTC  as we have Hike Difference
df_load_features.drop("Percentage hike offered (CTC)", axis = 1, inplace=True) #Drop Percentage hike offered (CTC)  as we have Hike Difference
df_load_features.drop("Percentage hike (CTC) expected ", axis = 1, inplace=True) #Drop Percentage hike (CTC) expected   as we have Hike Difference
df_load_features.drop("Date of last Interview", axis = 1, inplace=True) #Date of last Interview expected   as we have elapsed days for interview
df_load_features.drop("Date of first Interview", axis = 1, inplace=True) #Date of first Interview expected   as we have elapsed days for interview
df_load_features.drop("Written Test Date", axis = 1, inplace=True) #Date of first Interview expected   as we have elapsed days for interview
df_load_features.drop("DOB", axis = 1, inplace=True) #DOB doesnt seem to be relevant

print("\n\nHead data post transformation\n\n",df_load_features.head(20))
print("\ndata types of dataframe post transformation\n",df_load_features.dtypes)

df_load_features_raw_post_transform = df_load_features.copy() # Store transformed data in another dataframe for future use

##### Convert to Dummy Variable #####
'''
df_load_features = pd.get_dummies(df_load_features, columns=["Offered band","LOB","Joining location","Highest Educational Degree",
                                                 "Candidate source","Interview types","Interview Types2","Location"])
''' 
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
df_load_features["Offered band_cat"] = labelencoder.fit_transform(df_load_features['Offered band'])
df_load_features["LOB_cat"] = labelencoder.fit_transform(df_load_features['LOB'])
df_load_features["Joining location_cat"] = labelencoder.fit_transform(df_load_features['Joining location'])
df_load_features["Highest Educational Degree_cat"] = labelencoder.fit_transform(df_load_features['Highest Educational Degree'])
df_load_features["Candidate source_cat"] = labelencoder.fit_transform(df_load_features['Candidate source'])
df_load_features["Interview types_cat"] = labelencoder.fit_transform(df_load_features['Interview types'])
df_load_features["Interview Types2_cat"] = labelencoder.fit_transform(df_load_features['Interview Types2'])
df_load_features["Location_cat"] = labelencoder.fit_transform(df_load_features['Location'])



df_load_features.loc[:, 'Candidate relocation status'] = df_load_features.loc[:, 'Candidate relocation status'].map({
    'Yes':1,'No':0})
df_load_features.loc[:, 'Acceptance status'] = df_load_features.loc[:, 'Acceptance status'].map({
    'Joined':0,'Offer Rejected':1})
df_load_features.loc[:, 'Gender'] = df_load_features.loc[:, 'Gender'].map({'F': 0 ,'M': 1})

###### Dropping Encoded columns ######

df_load_features.drop("Offered band", axis = 1, inplace=True)
df_load_features.drop("LOB", axis = 1, inplace=True)
df_load_features.drop("Joining location", axis = 1, inplace=True)
df_load_features.drop("Highest Educational Degree", axis = 1, inplace=True)
df_load_features.drop("Candidate source", axis = 1, inplace=True)
df_load_features.drop("Interview types", axis = 1, inplace=True)
df_load_features.drop("Interview Types2", axis = 1, inplace=True)
df_load_features.drop("Location", axis = 1, inplace=True)

print("\n\nHead data post dummy Variable\n\n",df_load_features.head(5))
print("\ndata types of dataframe post dummy variable\n",df_load_features.dtypes)

##### Normalize and Standardize values #####

scaler = StandardScaler()
df_load_features[["Joining bonus amount","Notice period", "Total experience","Relevant experience","Year of graduation","Number of precious jobs"
                                      ,"Interview rounds","First interview duration","Last interaction duration"
                                                    ,"ElapsedDays_Application_Joining","Hike Difference","ElapsedDays_InterviewDates"
                                                    ]]  = scaler.fit_transform(df_load_features[["Joining bonus amount","Notice period", "Total experience","Relevant experience","Year of graduation","Number of precious jobs"
                                      ,"Interview rounds","First interview duration","Last interaction duration"
                                                    ,"ElapsedDays_Application_Joining","Hike Difference","ElapsedDays_InterviewDates"
                                                    ]])

print("\n\nHead data post scaling\n\n",df_load_features.head(5))
######## Finding outliers for numerical variables against Acceptance Status #########


tmp_df = df_load_features_raw_post_transform.loc[:, ["Joining bonus amount","Notice period", "Total experience","Relevant experience","Year of graduation","Number of precious jobs"
                                      ,"Interview rounds","First interview duration","Last interaction duration"
                                                    ,"ElapsedDays_Application_Joining","Hike Difference","ElapsedDays_InterviewDates"
                                                    ,"Acceptance status"]]
for column in tmp_df:
    if column!='Acceptance status':
        plt.figure(figsize=(10, 8))
        plt.title('{} OUTLIER DISTRIBUTION againt Acceptance Status'.format(column), fontsize=16)
        sn.boxplot(x="Acceptance status",y=column,data=tmp_df)
        plt.show()

```

```python
#######################################################################################################
#                                   PCA Analysis using library                                        #
#######################################################################################################

from sklearn.decomposition import PCA

####### FInal manipulation of data #######

df_load_features_copy = df_load_features.copy() 
df_load_acceptance = df_load_features_copy.loc[:, ['Acceptance status']]
df_load_features_copy.drop("Acceptance status", axis = 1, inplace=True) #Acceptance status is not required

pca = PCA(.95) # Here we can also give the percentage as a paramter to the PCA function as pca = PCA(.95). .95 means that we want to include 95% of the variance. Hence PCA will return the no of components which describe 95% of the variance. However we know from above computation that 2 components are enough so we have passed the 2 components.
principalComponents = pca.fit_transform(df_load_features_copy) 
principalDf = pd.DataFrame(data = principalComponents) 
print(principalDf.head(5)) # prints the top 5 rows


df_load_features_copy = np.array(principalDf)
df_load_acceptance = np.array(df_load_acceptance)


xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_copy, 
                                                df_load_acceptance.reshape(df_load_acceptance.shape[0],)
                                                , test_size = 0.2, random_state = 0) #split the datasset into train and validation
                                       
#print(xTrain[:5,:])
```

```python
#######################################################################################################
#                               PCA Analysis using custom implementation                              #
#######################################################################################################

df_load_features_forPCA = df_load_features.copy()
df_load_acceptance = df_load_features.loc[:, ['Acceptance status']] #Store acceptance status as Y 
df_load_features_forPCA.drop("Acceptance status", axis = 1, inplace=True) #Acceptance status is not required

cov_mat= np.cov(df_load_features_forPCA, rowvar=False) #creating the covariance matrix 
print("cov-Matrix-shape",cov_mat.shape)
eig_vals, eig_vecs = np.linalg.eig(cov_mat) #getting the eigen vectors and values 

#Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
#print(type(eig_pairs))
#Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()
print("\n",eig_pairs)
#Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('\n\n\nEigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])

tot = sum(eig_vals)
print("\n",tot)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
print("\n\n1. Variance Explained\n",var_exp)
cum_var_exp = np.cumsum(var_exp)
print("\n\n2. Cumulative Variance Explained\n",cum_var_exp)
print("\n\n3. Percentage of variance the first two principal components each contain\n ",var_exp[0:2])
print("\n\n4. Percentage of variance the first two principal components together contain\n",sum(var_exp[0:2]))

#print(eig_pairs[0][1])
#print(eig_pairs[1][1])

res_list = [i for i, value in enumerate(cum_var_exp) if value > 95] 
index_with_req_per = (res_list[0])
print("no of values which gives more than 95% information",index_with_req_per+1)


###### Create projection matrix ######

matrix_w = eig_pairs[0][1].reshape(22,1)
for eigenindex in range(1,index_with_req_per+1):
    matrix_w = np.hstack((matrix_w,eig_pairs[eigenindex][1].reshape(22,1)))
    

print('Matrix W:\n', matrix_w.shape)

###### Create final component matrix ######

FinalComponentMatrix = df_load_features_forPCA.dot(matrix_w)
FinalComponentMatrix = np.array(FinalComponentMatrix)
df_load_acceptance = np.array(df_load_acceptance)
print(FinalComponentMatrix)


xTrain, xTest, yTrain, yTest = train_test_split(FinalComponentMatrix, 
                                                df_load_acceptance.reshape(df_load_acceptance.shape[0],)
                                                , test_size = 0.2, random_state = 0) #split the datasset into train and validation
```

```python
#######################################################################################################
#                        Create & Test Model for coded component matrix                               #
#######################################################################################################

from sklearn.metrics import confusion_matrix
logreg = LogisticRegression()
logreg.fit(xTrain, yTrain)
y_pred = logreg.predict(xTest)
print('Accuracy of logistic regression classifier on test set using coded component matrix: {:.2f}'.format(logreg.score(xTest, yTest)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
```

```python
#######################################################################################################
#                          Forward Feature Selection Method & Model                                   #
#######################################################################################################

from sklearn.metrics import confusion_matrix

df_load_features_for_ForwardFeature = df_load_features.copy()
df_load_acceptance = df_load_features.loc[:, ['Acceptance status']] #Store acceptance status as Y 
df_load_features_for_ForwardFeature.drop("Acceptance status", axis = 1, inplace=True) #Acceptance status is not required

#Define Sequential Forward Selection (sfs)
sfs = SFS(LogisticRegression(),
           k_features=5,
           forward=True,
           floating=False,
           scoring = 'r2',
           cv = 0)


#Use SFS to select the top 5 features 
sfs.fit(df_load_features_for_ForwardFeature, df_load_acceptance)

#Create a dataframe for the SFS results 
df_SFS_results = pd.DataFrame(sfs.subsets_).transpose()
print("Most important features as per forward selection:\n")
print(df_SFS_results)
feat_cols = list(sfs.k_feature_idx_)


df_load_features_for_ForwardFeature = np.array(df_load_features_for_ForwardFeature)
df_load_acceptance = np.array(df_load_acceptance)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_ForwardFeature, 
                                                df_load_acceptance.reshape(df_load_acceptance.shape[0],)
                                                , test_size = 0.2, random_state = 0) #split the datasset into train and validation
logreg = LogisticRegression()
logreg.fit(xTrain[:,feat_cols], yTrain) #selecting top 5 features identified by forward selection
y_pred = logreg.predict(xTest[:,feat_cols])
print('Accuracy of logistic regression classifier on test set using forward feature elimination: {:.2f}'.format(logreg.score(xTest[:,feat_cols], yTest)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
```

```python
#######################################################################################################
#                          Backward Feature Selection Method & Model                                  #
#######################################################################################################

from sklearn.metrics import confusion_matrix

df_load_features_for_BackwardFeature = df_load_features.copy()
df_load_acceptance = df_load_features.loc[:, ['Acceptance status']] #Store acceptance status as Y 
df_load_features_for_BackwardFeature.drop("Acceptance status", axis = 1, inplace=True) #Acceptance status is not required

#Define Backward Selection (rfe)
rfecv = RFECV(estimator=LogisticRegression(), step=1, cv=StratifiedKFold(10), scoring='accuracy')


#Use rfe to select the top 5 features 
rfecv.fit(df_load_features_for_BackwardFeature, df_load_acceptance)

#Print optimal features
print('Optimal number of features: {}'.format(rfecv.n_features_))

#Drop non-optimal features
df_load_features_for_BackwardFeature.drop(df_load_features_for_BackwardFeature.columns[np.where(rfecv.support_ == False)[0]], axis=1, inplace=True)

df_load_features_for_BackwardFeature = np.array(df_load_features_for_BackwardFeature)
df_load_acceptance = np.array(df_load_acceptance)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_BackwardFeature, 
                                                df_load_acceptance.reshape(df_load_acceptance.shape[0],)
                                                , test_size = 0.2, random_state = 0) #split the datasset into train and validation
logreg = LogisticRegression()
logreg.fit(xTrain, yTrain) #selecting top 5 features identified by forward selection
y_pred = logreg.predict(xTest)
print('Accuracy of logistic regression classifier on test set using backward feature elimination: {:.2f}'.format(logreg.score(xTest, yTest)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
```

```python
#######################################################################################################
#                          Stepwise Feature Selection Method & Model                                  #
#######################################################################################################

from sklearn.metrics import confusion_matrix

df_load_features_for_StepwiseForwardFeature = df_load_features.copy()
df_load_acceptance = df_load_features.loc[:, ['Acceptance status']] #Store acceptance status as Y 
df_load_features_for_StepwiseForwardFeature.drop("Acceptance status", axis = 1, inplace=True) #Acceptance status is not required

#Define Sequential Forward Selection (sfs)
sfs = SFS(LogisticRegression(),
           k_features=5,
           forward=True,
           floating=True,
           scoring = 'r2',
           cv = 0)


#Use SFS to select the top 5 features 
sfs.fit(df_load_features_for_StepwiseForwardFeature, df_load_acceptance)

#Create a dataframe for the SFS results 
df_SFS_results = pd.DataFrame(sfs.subsets_).transpose()
print("Most important features as per stepwise forward selection:\n")
print(df_SFS_results)
feat_cols = list(sfs.k_feature_idx_)


df_load_features_for_StepwiseForwardFeature = np.array(df_load_features_for_StepwiseForwardFeature)
df_load_acceptance = np.array(df_load_acceptance)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_StepwiseForwardFeature, 
                                                df_load_acceptance.reshape(df_load_acceptance.shape[0],)
                                                , test_size = 0.2, random_state = 0) #split the datasset into train and validation
logreg = LogisticRegression()
logreg.fit(xTrain[:,feat_cols], yTrain) #selecting top 5 features identified by forward selection
y_pred = logreg.predict(xTest[:,feat_cols])
print('Accuracy of logistic regression classifier on test set using stepwise forward feature selections: {:.2f}'.format(logreg.score(xTest[:,feat_cols], yTest)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
```
