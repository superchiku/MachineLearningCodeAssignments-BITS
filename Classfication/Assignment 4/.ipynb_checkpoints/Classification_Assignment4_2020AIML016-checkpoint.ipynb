```python
# find and remove correlated features with a given threshold
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = df_final_data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

# <font color=Red>(C) Split the Data

1. Divide into features and target 
2. Divide into training and testing 
3. Ensure that data is divided with a more or less equal ratio of class 1 and 0 in both test and train set   </font>

<font color = Green>
    We will use stratified splitting to divide the data into equal ratio of classes
</font>








```python
# Create accuracy score & model dict arrays to draw graph later #
accscore_arr = []
aucscore_arr = []
accscore_model = []


# seperate X,Z and y variables
y = df_final_data['not.fully.paid']
X = df_final_data.drop(['not.fully.paid'],axis=1)

#Scaling the date before splitting into train and test

scaler = StandardScaler()
X[["purpose","int.rate", "installment","log.annual.inc","dti","fico"
                                      ,"days.with.cr.line","revol.bal","revol.util"
                                                    ,"inq.last.6mths","delinq.2yrs"
                                                    ]]  = scaler.fit_transform(X[["purpose","int.rate", "installment","log.annual.inc","dti","fico"
                                      ,"days.with.cr.line","revol.bal","revol.util"
                                                    ,"inq.last.6mths","delinq.2yrs"
                                                    ]])

print("\n\nHead data post scaling\n\n",X.head(5))

X=np.array(X)
y=np.array(y)
X.shape

# Splitting the dataset into the Training set and Test set using stratification 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1 , stratify=y)
```

# <font color=Red>(D)Implement the SVM Model

1. Scale the features 
2. Train Model 
3. Predict  </font>



# Testing Model with all features without tuning

```python
# fit SVM to the training set 

from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import scikitplot as skplt
#Create SVM Classifier
clf = SVC()

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
#y_pred_prob = clf.predict_proba(X_test)

confusion_matrix = confusion_matrix(y_test, y_pred)
acc_score = accuracy_score(y_test, y_pred)
accscore_arr.append(acc_score)
accscore_model.append('All Features SVM with no tuning')
print("accuracy score for non tuned SVM  with all features: \n",confusion_matrix)
print ("accuracy score = ",round(acc_score*100,2),"%")

# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",recall_score(y_test, y_pred))

########## Drawing the AUC curves ###########
#y_pred_proba = y_pred[:,1]
y_pred_proba = y_pred
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
aucscore_arr.append(auc)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label=" auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

# HyperParameter tuning for SVM with all features

```python
# importing RepeatedStratifiedKFold
#from sklearn.model_selection import RepeatedStratifiedKFold
# define evaluation
#cv = RepeatedStratifiedKFold(n_splits=2, n_repeats=1, random_state=1)

##### Stratified KFOLD validation was taking very long to train , so going for a simple split in a stratified manner #####
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1 , stratify=y)
```

```python
# fit SVM  to the training set 

from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import scikitplot as skplt
#Create SVM Classifier
clf = SVC(kernel = 'sigmoid', degree=20, C=10)

#grid_values = {'penalty': ['l1', 'l2'],'C':[0.001,.009,0.01,.09,1,5,10,25]}
#grid_values = {'penalty': ['l1', 'l2']}
# define search space
'''
space = dict()
space['C'] = [0.1, 1, 10, 100, 1000]
space['gamma'] = [1, 0.1, 0.01, 0.001, 0.0001]
space['kernel'] = ['linear','poly','rbf']
'''
#grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3) 
#grid_clf_acc = GridSearchCV(clf, space,scoring = 'accuracy' ,cv=cv, n_jobs=-1,verbose = 3 , refit = True)


#Train the model using the training sets
result = clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
#y_pred_prob = grid_clf_acc.predict_proba(X_test)

confusion_matrix = confusion_matrix(y_test, y_pred)
acc_score = accuracy_score(y_test, y_pred)
accscore_arr.append(acc_score)
accscore_model.append('Hyper Parameter tuned SVM with all features')

#print('Best Hyperparameters: %s' % result.best_params_)
print("accuracy score for SVM with all features and hyperparameter defined \n",confusion_matrix)
print ("accuracy score = ",round(acc_score*100,2),"%")
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",recall_score(y_test, y_pred))

########## Drawing the AUC curves ###########
#y_pred_proba = y_pred_prob[:,1]
y_pred_proba = y_pred
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
aucscore_arr.append(auc)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label=" auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
