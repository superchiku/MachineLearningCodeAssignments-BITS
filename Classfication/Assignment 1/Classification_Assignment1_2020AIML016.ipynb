# <font color=Red>(D)FEATURE ENGINEERING & MODEL BUILDINGAND DIAGNOSTICS </font>

<font color=green>FEATURE ENGINEERING IS AN IMPORTANT STEP TO DEVELOP AND IMPROVE PERFORMANCE OF MACHINE LEARNING MODELS. THESE TECHNIQUES CAN ALSO BE USED TO IDENTIFY THE VARIABLES THAT IMPACTS THE OUTCOME OF THE MODEL. ITS BASICALLY PROCESS OF IDENTIFYING AND EXTRACTING THE USEFUL FEATURES FROM THE AVAILABLE DATA. THE PRIMARY GOAL IS TO DERIVE A SET OF FEATURES THAT BEST REPRESENT THE INSIGHTS HIDDEN IN THE DATA, WITH A SIMPLER MODEL THAT GENERALIZES WELL TO FUTURE (UNKNOWN) OBSERVATIONS.
</font>


MODEL BUILDINGAND DIAGNOSTICS:-


IN THIS STAGE FIRST DATA IS DIVIDED INTO TRAIN AND TEST DATA. THE SUBSET CAN ALSO BE CREATED USING RANDOM / STRATIFIED SAMPLING PROCEDURE. THIS IS IMPORTANT STEP TO MEASEURE THE PERFORMANCE OF MODEL USING DATASET NOT USED IN MODEL BUILDING. IT IS ALSO ESSENTIAL TO CHECK FOR ANY OVERFITTING OF THE MODEL. THEN MODEL IS BUILT USING TRAINING DATASET TO ESTIMATE THE MODEL PARAMETERS. THE METHOD OF CLASSIFICATIONS CAN BE UTILIZED FOR THE SAME.





```python
# find and remove correlated features with a given threshold
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = df_final_data_1.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

# analysis with Forward Feature Selection 

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
#######################################################################################################
#                          Forward Feature Selection Method & Model                                   #
#######################################################################################################

from sklearn.metrics import confusion_matrix

df_load_features_for_ForwardFeature = df_final_data_2.copy()
df_load_Outcome = df_final_data_2.loc[:, ['Outcome']] #Store Outcome as Y 
df_load_features_for_ForwardFeature.drop("Outcome", axis = 1, inplace=True) #Outcome is not required

#Define Sequential Forward Selection (sfs)
sfs = SFS( KNeighborsClassifier(n_neighbors=3),
           k_features=4,
           forward=True,
           floating=False,
           scoring = 'accuracy',
           cv = 0)


#Use SFS to select the top 5 features 
sfs.fit(df_load_features_for_ForwardFeature, df_load_Outcome)

#Create a dataframe for the SFS results 
df_SFS_results = pd.DataFrame(sfs.subsets_).transpose()
print("Most important features as per forward selection:\n")
print(df_SFS_results)
feat_cols = list(sfs.k_feature_idx_)


df_load_features_for_ForwardFeature = np.array(df_load_features_for_ForwardFeature)
df_load_Outcome = np.array(df_load_Outcome)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_ForwardFeature, 
                                                df_load_Outcome.reshape(df_load_Outcome.shape[0],)
                                                , test_size = 0.25, random_state = 0) #split the datasset into train and validation
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xTrain[:,feat_cols], yTrain) #selecting top 5 features identified by forward selection
y_pred = knn.predict(xTest[:,feat_cols])
acc_score = accuracy_score(y_test, y_pred)
print('Accuracy of Knn classifier on test set using forward feature elimination: {:.2f} %'.format(round(acc_score*100,2)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
```

# PCA Analysis using library  

```python
#######################################################################################################
#                                   PCA Analysis using library                                        #
#######################################################################################################

from sklearn.decomposition import PCA

####### FInal manipulation of data #######

df_load_features_copy = df_final_data_2.copy()
df_load_Outcome = df_final_data_2.loc[:, ['Outcome']] #Store Outcome as Y 
df_load_features_copy.drop("Outcome", axis = 1, inplace=True) #Outcome is not required

pca = PCA(.95) # Here we can also give the percentage as a paramter to the PCA function as pca = PCA(.95). .95 means that we want to include 95% of the variance. Hence PCA will return the no of components which describe 95% of the variance.
principalComponents = pca.fit_transform(df_load_features_copy) 
principalDf = pd.DataFrame(data = principalComponents) 
print(principalDf.head(5)) # prints the top 5 rows


df_load_features_copy = np.array(principalDf)
df_load_Outcome = np.array(df_load_Outcome)


xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_copy, 
                                                df_load_Outcome.reshape(df_load_Outcome.shape[0],)
                                                , test_size = 0.25, random_state = 0) #split the datasset into train and validation
                                       

#######################################################################################################
#                        Create & Test Model for library component matrix                             #
#######################################################################################################
from sklearn.metrics import confusion_matrix

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xTrain, yTrain) #selecting top 5 features identified by forward selection
y_pred = knn.predict(xTest)
acc_score = accuracy_score(y_test, y_pred)
print('Accuracy of Knn classifier on test set using forward feature elimination: {:.2f} %'.format(round(acc_score*100,2)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)
    #print(xTrain[:5,:])
```

# Using Elbow Method to find the optimal value of K

```python
df_load_features_for_Elbow = df_final_data_2.copy()
df_load_features_for_Elbow = df_load_features_for_Elbow.loc[:, ["Pregnancies","Glucose","BMI","Age"]]
df_load_Outcome = df_final_data_2.loc[:, ['Outcome']] #Store Outcome as Y 


df_load_features_for_Elbow = np.array(df_load_features_for_Elbow)
df_load_Outcome = np.array(df_load_Outcome)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_Elbow, 
                                                df_load_Outcome.reshape(df_load_Outcome.shape[0],)
                                                , test_size = 0.25, random_state = 0) #split the datasset into train and validation

accscore_arr = []
# Might take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xTrain, yTrain)
    y_pred = knn.predict(xTest)
    acc_score = accuracy_score(y_test, y_pred)
    accscore_arr.append(acc_score)
    

plt.figure(figsize=(10,6))
plt.plot(range(1,40),accscore_arr,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
```

# <font color=Red>From the above we can determine that best accuracy is obtained for K value =8 post which it starts decreasing</font>

# Best model performance with K = 8 below

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import scikitplot as skplt

df_load_features_for_Elbow = df_final_data_2.copy()
df_load_features_for_Elbow = df_load_features_for_Elbow.loc[:, ["Pregnancies","Glucose","BMI","Age"]]
df_load_Outcome = df_final_data_2.loc[:, ['Outcome']] #Store Outcome as Y 


df_load_features_for_Elbow = np.array(df_load_features_for_Elbow)
df_load_Outcome = np.array(df_load_Outcome)
xTrain, xTest, yTrain, yTest = train_test_split(df_load_features_for_Elbow, 
                                                df_load_Outcome.reshape(df_load_Outcome.shape[0],)
                                                , test_size = 0.25, random_state = 0) #split the datasset into train and validation

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(xTrain, yTrain) #selecting top 5 features identified by forward selection
y_pred = knn.predict(xTest)
acc_score = accuracy_score(y_test, y_pred)
print('Best accuracy of Knn classifier on test set using elbow method and highly correlated columns from filter method: {:.2f} %'.format(round(acc_score*100,2)))

confusion_matrix = confusion_matrix(yTest, y_pred)
print("confusion matrix\n",confusion_matrix)

########## Drawing the AUC curves ###########
y_pred_proba = y_pred
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label=" auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
